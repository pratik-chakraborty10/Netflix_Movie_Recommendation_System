## Importing libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing dataset

```
dataset=pd.read_csv(" Netflix_Movie_Recommendation.csv",header=None)
transactions=[]

for i in range(0,7466):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```
dataset.describe()
```

## Apriori Training on Dataset

```
from apyori import apriori
#ECLAT algo used here only support required.
#Eclat (Equivalence Class Clustering and Bottom-Up Lattice Traversal) is another popular algorithm used for frequent item set mining in transactional databases. Similar to the Apriori algorithm, Eclat is used to find sets of items that frequently occur together in transactions.
movie_intelligence = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualizing

### Raw Results

```
results=list(movie_intelligence)
```

```
results
```

### Proper Display

```
def inspect(results):
    Movie1 = [tuple(result[2][0][0])[0] for result in results]
    Movie2 = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]

    return list(zip(Movie1, Movie2, supports))

# Assuming results contain the association rules generated
results_intelligence = pd.DataFrame(inspect(results), columns=['Movie1', 'Movie2', 'Support'])
```

```
results_intelligence
```

```
results_intelligence.nlargest(n=10,columns="Support")
```
